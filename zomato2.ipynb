```python
rating=fin_dat.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0: 'Racting count'})
```

```python
rating
```

1.When Rating is between 4.5 to 4.9 --> Excellent

3.When Rating is between 3.5 to 3.0--> Good

4.When Rating is between 2.5 to 2.9 -->Average

5.When Rating is between 2.0 to 2.4 -->Poor

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x="Aggregate rating",y="Racting count",data=rating)
```

```python
sns.barplot(x="Aggregate rating",y="Racting count",hue= 'Rating color', data=rating, palette= ['white','red','orange','yellow','green','green'])
```

Observation :

1. Not rated count is very high

2.Maximum number of rating are between 2.5 to 3.4

```python
#count plot
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
```

Now we can Observe that blue is having less records were Red having highest Records
